```python
#support vector machine
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("svm_data.csv")
df
```

```python
df[df.Target==1].head()
```

```python
df0 = df[df.Target==0]
df1 = df[df.Target==1]
```

```python
plt.scatter(df0["Feature_1"], df0["Feature_2"], color="green", marker = "+")
plt.scatter(df1["Feature_1"], df1["Feature_2"], color="blue", marker = ".")
plt.xlabel("Feature_1")
plt.ylabel("Feature_2")
```

```python
x = df.drop(["Target"], axis = "columns")
x
```

```python
y = df.Target
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state = 0)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
#trying other parameters
model_c = SVC(C=100)
model_gamma = SVC(gamma=120)
model_kernel = SVC(kernel="linear")
```
