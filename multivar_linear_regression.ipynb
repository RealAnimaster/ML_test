```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('sample_multifeature_data.csv')
df
```

```python
#split df into feature/independant and target/dependant
x = df.iloc[:,:-1].values
x[0:5]
```

```python
y = df.iloc[:,1].values
y[0:5]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
error = y_pred - y_test
error
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
r2
```

```python
class CustomLinearRegression:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.w = None
        self.b = None

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        
        # Check for NaN or Inf values
        if np.isnan(X).any() or np.isnan(y).any():
            raise ValueError("Input contains NaN values.")
        if np.isinf(X).any() or np.isinf(y).any():
            raise ValueError("Input contains infinity values.")
        
        # Add intercept column (ones) for the bias term
        X = np.c_[np.ones(X.shape[0]), X]  # Adding a column of ones for the bias term
        
        # Closed-form solution: w = (X^T * X)^-1 * X^T * y
        self.w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
        self.b = self.w[0]
        self.w = self.w[1:]

    def predict(self, X):
        # Convert X to a NumPy array
        X = np.array(X)

        # Check for invalid inputs
        if np.isnan(X).any():
            raise ValueError("Input contains NaN values. Please clean your data.")
        if np.isinf(X).any():
            raise ValueError("Input contains infinity values. Please clean your data.")

        # Prediction
        return np.dot(X, self.w) + self.b  # Linear prediction for multiple features
```

```python
custom_model = CustomLinearRegression(learning_rate=0.01, epochs=1000)
custom_model.fit(x_train, y_train)
```

```python
custom_y_pred = custom_model.predict(x_test)
```

```python
custom_error = custom_y_pred - y_test
custom_error
```

```python
compare_error = error-custom_error
compare_error
```
