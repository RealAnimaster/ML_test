```python
#random forest
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("random_forest_data.csv")
df
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(["Target"], axis = "columns"), df.Target, train_size = 0.8, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
#fine tuning major parameters with randomized search
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

# Define the parameter grid
param_dist = {
    "n_estimators": range(10, 240, 10),          # Number of trees
    "max_depth": [None, 10, 20, 30, 40, 50],    # Tree depth
    "min_samples_split": [2, 5, 10],            # Minimum samples to split
    "min_samples_leaf": [1, 2, 4],              # Minimum samples in leaf
    "max_features": ["sqrt", "log2", None],     # Features considered at each split
    "bootstrap": [True, False]                  # Bootstrap sampling
}

# RandomizedSearchCV
rf = RandomForestClassifier(random_state=42)
random_search = RandomizedSearchCV(
    estimator=rf, 
    param_distributions=param_dist,
    n_iter=100,          # Number of random combinations to try
    cv=3,                # 3-fold cross-validation
    verbose=2,           # Print progress
    random_state=42,
    n_jobs=-1            # Parallelism
)

# Fit the model
random_search.fit(x_train, y_train)

# Best parameters and score
print("Best parameters:", random_search.best_params_)
print("Best score:", random_search.best_score_)
```

```python
#refining with grid search for precision
from sklearn.model_selection import GridSearchCV

# Define the refined parameter grid
param_grid = {
    "n_estimators": [120, 140, 160],       # Refined around the best n_estimators
    "max_depth": [15, 20, 25],            # Refined around the best max_depth
    "min_samples_split": [2],             # Keep as is (best value)
    "min_samples_leaf": [1, 2, 3],        # Test nearby values
    "max_features": ["sqrt", "log2"],     # Test nearby options
    "bootstrap": [False]                  # Keep as is (best value)
}

# Initialize the RandomForestClassifier
rf = RandomForestClassifier(random_state=42)

# GridSearchCV for fine-tuning
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,               # 3-fold cross-validation
    verbose=2,          # Print progress
    n_jobs=-1           # Use all available cores
)

# Fit the model with the refined grid
grid_search.fit(x_train, y_train)

# Output the best parameters and score
print("Best parameters from grid search:", grid_search.best_params_)
print("Best score from grid search:", grid_search.best_score_)
```

```python
#comparing both models
# Randomized search model
random_model = RandomForestClassifier(**random_search.best_params_, random_state=42)
random_model.fit(x_train, y_train)
random_score = random_model.score(x_test, y_test)

# Grid search model
grid_model = RandomForestClassifier(**grid_search.best_params_, random_state=42)
grid_model.fit(x_train, y_train)
grid_score = grid_model.score(x_test, y_test)

print(f"Randomized Search Test Score: {random_score:.2f}")
print(f"Grid Search Test Score: {grid_score:.2f}")
```

```python
#insights
#outcome suggests model performs equally well with different configs, so it's relatively robust
#max_depth is reduced from 20 to 15 but the score remains same, maybe indicates model is not benefitting from deeper trees, potentially avoids over fitting
#smaller depth reduces model complexity
#computational cost is reduced while maintaining performance
#model is likely relying on ensemble power (multiple trees) than depth of individual trees
```

```python
# Final model with grid search parameters
final_model = RandomForestClassifier(
    **grid_search.best_params_,
    random_state=42
)
final_model.fit(x_train, y_train)

# Test the final model
final_score = final_model.score(x_test, y_test)
print(f"Final Model Test Score: {final_score:.2f}")
```
